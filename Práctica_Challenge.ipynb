# 4. Productos más y menos vendidos

```python
import numpy as np
import matplotlib.pyplot as plt

etiquetas = ['Tienda 1', 'Tienda 2','Tienda 3','Tienda 4']
valores = [60, 65, 57, 62]
valores2 = [33, 32, 35, 33]

co = np.arange(len(valores))
an = 0.30

fig, ax = plt.subplots()
ax.bar(co - an/2, valores, an, label = 'Más vendidos', color='skyblue')
ax.bar(co + an/2, valores2, an, label = 'Menos vendidos', color='orange')

for i,j in zip(co, valores):
  ax.annotate(j, xy = (i -0.2, j + 0.2))

for i,j in zip(co, valores2):
  ax.annotate(j, xy = (i +0.1, j + 0.2))

ax.set_title('Productos más y menos vendidos')
ax.set_ylabel('Cantidad de productos vendidos')
ax.set_xticks(co)
ax.set_xticklabels(etiquetas)
plt.legend()
plt.show()
```
